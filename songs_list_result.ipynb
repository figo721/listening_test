```python
import os
import gzip
import json
from pathlib import Path
    
my_dir = os.getcwd()
def _stream_lines(path: Path = Path(my_dir + '/songs.json.gz')):
    with gzip.open(path, 'r') as zipped_file:
        for line in zipped_file:
            yield json.loads(line)


data = list(_stream_lines())
```

```python
for elem in data:
    print(elem)
```

```python
import pandas as pd

df = pd.DataFrame.from_dict(data)
```

```python
df_dropped_by_uri = df.drop_duplicates(subset=['uri'])
```

```python
df_dropped_by_uri = df_dropped_by_uri.reset_index()
```
